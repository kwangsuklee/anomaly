```python
# function to reshape features into (samples, time steps, features) 
def gen_sequence(id_df, seq_length, seq_cols):
    """ Only sequences that meet the window-length are considered, no padding is used. This means for testing
    we need to drop those which are below the window-length. An alternative would be to pad sequences so that
    we can use shorter ones """
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        yield data_array[start:stop, :]
```

```python
# pick the feature columns 
sensor_cols = ['s' + str(i) for i in range(1,22)]
sequence_cols = ['setting1', 'setting2', 'setting3', 'cycle_norm']
sequence_cols.extend(sensor_cols)
```

```python
# generator for the sequences
seq_gen = (list(gen_sequence(train_df[train_df['id']==id], sequence_length, sequence_cols)) 
           for id in train_df['id'].unique())
```

```python
# generate sequences and convert to numpy array
seq_array = np.concatenate(list(seq_gen)).astype(np.float32)
seq_array.shape
```

```python
# function to generate labels
def gen_labels(id_df, seq_length, label):
    data_array = id_df[label].values
    num_elements = data_array.shape[0]
    return data_array[seq_length:num_elements, :]
```

```python
# generate labels
label_gen = [gen_labels(train_df[train_df['id']==id], sequence_length, ['RUL']) 
             for id in train_df['id'].unique()]
label_array = np.concatenate(label_gen).astype(np.float32)
label_array.shape
```

```python
#from keras.layers import Bidirectional
from keras.models import Sequential, Model 
from keras.layers import Dense, Input, Flatten, Add, concatenate, Dropout
from keras.layers import LSTM, Bidirectional 
from keras.layers.convolutional import Conv1D
```

```python
# build the network
nb_features = seq_array.shape[2]
nb_out = label_array.shape[1]

"""model = Sequential()
# model.add(Bidirectional(LSTM(units=100, return_sequences=True), input_shape=(sequence_length, nb_features)))
model.add(LSTM(
         units=100,
         return_sequences=True,
         input_shape=(sequence_length, nb_features)))
model.add(Dropout(0.4))

model.add(LSTM(
          units=100,
          return_sequences=False))
model.add(Dropout(0.4))

model.add(Dense(units=1, activation='relu'))
model.add(Activation("relu"))

model.compile(loss="mse", optimizer="rmsprop", metrics=['mse'])
"""

# sequence_length = 50   # Original 

# sequence_length = 100  # New timesteps 

visible1 = Input(shape=(sequence_length, nb_features))    # 100, 25,     50 타임스텝, 25 피처 

#  25,   conv_11 = Conv1D(filters=25, kernel_size=1, strides=1, activation='relu')(visible1)
conv_11 = Conv1D(filters=200, kernel_size=1, strides=1, activation='relu')(visible1)
conv_11_out = Dropout(0.5)(conv_11)

conv_12 = Conv1D(filters=25, kernel_size=1, strides=1, activation='relu')(conv_11_out)
conv_12_out = Dropout(0.5)(conv_12)

residual_1 = keras.layers.Add()([visible1, conv_12_out])   #  100, 25  (50, 25) (50, 50===>25)

#  25,   ls11 = (LSTM(25, return_sequences=True))(residual_1) 
ls11 = (LSTM(400, return_sequences=True))(residual_1) 
ls11_out = Dropout(0.5)(ls11)  # Dropout 0.2 -> 0.5 

ls12 = (LSTM(25, return_sequences=True))(ls11_out) 
ls12_out = Dropout(0.5)(ls12)

residual_2 = keras.layers.Add()([residual_1, ls12_out])   #  100, 25  (50, 25) (50, 50===>25)
    
ls21 = (LSTM(200, return_sequences=True))(residual_2) 
ls21_out = Dropout(0.5)(ls21)  # Dropout 0.2 -> 0.5 

ls22 = (LSTM(25, return_sequences=True))(ls21_out)       
ls22_out = Dropout(0.5)(ls22)

residual_3 = keras.layers.Add()([residual_2, ls22_out])   #  100, 25  (50, 25) (50, 50===>25)

ls31 = (LSTM(100, return_sequences=True))(residual_3) 
ls31_out = Dropout(0.5)(ls31)  # Dropout 0.2 -> 0.5 

ls32 = (LSTM(25, return_sequences=True))(ls31_out)        
ls32_out = Dropout(0.5)(ls32)

residual_4 = keras.layers.Add()([residual_3, ls32_out])   #  100, 25  (50, 25) (50, 50===>25)

bi11 = (Bidirectional(LSTM(100, return_sequences=True)))(residual_4) # 50 
bi11_out = Dropout(0.5)(bi11)

bi12 = (Bidirectional(LSTM(200)))(bi11_out)     # 100,   Final LSTM, do not return sequence
bi12_out = Dropout(0.5)(bi12)

out1 = Dense(units=nb_out, activation='relu')(bi12_out)    # Dense(1, ), nb_out = 1

model = Model(inputs=[visible1], outputs=[out1]) # from keras.models import Model

#model.compile(loss="mse", optimizer="rmsprop", metrics=['mse'])
#model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse', 'mae', 'mape']) 
model.compile(loss='mean_squared_error', optimizer='rmsprop', metrics=['mse', 'mae', 'mape']) 
```

```python
print(model.summary())
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

STAMP = 'predictive_regression_kalhman'
print(STAMP)

early_stopping =EarlyStopping(monitor='val_loss', patience=15)
bst_model_path = STAMP + '.h5'
model_checkpoint = ModelCheckpoint(bst_model_path, save_best_only=True, save_weights_only=True)
```

```python
%%time
# fit the network
hist = model.fit(seq_array, label_array, epochs=1000, batch_size=200, validation_split=0.05, verbose=1,callbacks=[early_stopping, model_checkpoint])
```
